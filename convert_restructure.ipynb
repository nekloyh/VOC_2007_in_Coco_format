```python
# dataset/
# ├── images/
# │   ├── train/
# │   │   ├── image1.jpg
# │   │   ├── image2.jpg
# │   │   └── ...
# │   ├── val/
# │   │   ├── image1.jpg
# │   │   ├── image2.jpg
# │   │   └── ...
# └── annotations/
#     ├── instances_train.json
#     ├── instances_val.json
#     └── ...

import os
import shutil

data_path = "D:\\university\\OJT202\\20250217_Object Detection\\PASCAL-VOC-2007-1"

source_images_train = os.path.join(data_path, "train")
source_images_val = os.path.join(data_path, "valid")
source_annotations_train = os.path.join(data_path, "train", "_annotations.coco.json")
source_annotations_val = os.path.join(data_path, "valid", "_annotations.coco.json")


destination_root = "D:\\university\\OJT202\\20250217_Object Detection\\PASCAL_VOC_2007_COCO_Format"

destination_images_train = os.path.join(destination_root, "images", "train")
destination_images_val = os.path.join(destination_root, "images", "val")
destination_annotations = os.path.join(destination_root, "annotations")


os.makedirs(destination_images_train, exist_ok=True)
os.makedirs(destination_images_val, exist_ok=True)
os.makedirs(destination_annotations, exist_ok=True)

for filename in os.listdir(source_images_train):
    if filename.endswith((".jpg")):
        source_path = os.path.join(source_images_train, filename)
        destination_path = os.path.join(destination_images_train, filename)
        shutil.move(source_path, destination_path)
for filename in os.listdir(source_images_val):
    if filename.endswith((".jpg")):
        source_path = os.path.join(source_images_val, filename)
        destination_path = os.path.join(destination_images_val, filename)
        shutil.move(source_path, destination_path)


shutil.copy(source_annotations_train, os.path.join(destination_annotations, "instances_train.json"))
shutil.copy(source_annotations_val, os.path.join(destination_annotations, "instances_val.json"))
```
